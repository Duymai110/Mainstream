```python
############################################################
class MovieNode:
    def __init__(self, title, year, director, genre):
        self.title = title
        self.year = year
        self.director = director
        self.genre = genre
        self.next = None
        self.prev = None

    def __str__(self):
        return f"Title:  {self.title:<25} Year: {self.year:<6} Director: {self.director:<20} Genre: {self.genre:<10}"

    def __lt__(self, other):
        if self.title < other.title:
            return True
        else:
            return False

############################################################
class MovieList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def __str__(self):
        rstr = ''
        curr = self.head
        while curr:
            rstr += curr.__str__() + '\n'
            curr = curr.next    
        return rstr

    # Append movie at end of MovieList
    def append_movie(self, new_movie): 
        if self.head == None:
            self.head = new_movie
            self.tail = new_movie
        else:
            self.tail.next = new_movie
            new_movie.prev = self.tail
            self.tail = new_movie
    
    # Insert new_movie after curr_movie
    def insert_movie_after(self, curr_movie, new_movie): 
        if self.head is None:
            self.head = new_movie
            self.tail = new_movie
        elif curr_movie is self.tail:
            self.tail.next = new_movie
            new_movie.prev = self.tail
            self.tail = new_movie
        else:
            succ = curr_movie.next
            new_movie.next = succ
            new_movie.prev = curr_movie
            curr_movie.next = new_movie
            succ.prev = new_movie
    
    # Delete current_movie from MovieList
    def delete_movie(self, current_movie):
        successor_node = current_movie.next
        predecessor_node = current_movie.prev

        if successor_node is not None:
            successor_node.prev = predecessor_node
        if predecessor_node is not None:
            predecessor_node.next = successor_node
        if current_movie is self.head:
            self.head = successor_node
        if current_movie is self.tail:
            self.tail = predecessor_node
    



#################################################
# Task 1: complete the following function
# remove_duplicates(m_list: MovieList) -> MovieList:
#################################################            
def remove_duplicates(movie_list):
    seen_titles = {}
    curr = movie_list.head
    
    while curr:
        if curr.title in seen_titles:
            prev = curr.prev
            next = curr.next
            prev.next = next
            if next:
                next.prev = prev
            curr = next
        else:
            seen_titles[curr.title] = True
            curr = curr.next
    
    return movie_list
            
    
#################################################
# Task 2: complete the following function
# addMovieSortedOrder(sorted_list: MovieList, movie: MovieNode) -> MovieList:
#################################################
def addMovieSortedOrder(sorted_list, movie):
    curr = sorted_list.head     
    flag = False
    if sorted_list.head is None:  # list is empty
        sorted_list.append_movie(movie)
        return sorted_list
    
    if movie.title.upper() <= sorted_list.head.title.upper():
        sorted_list.head.prev = movie
        movie.next = sorted_list.head
        sorted_list.head = sorted_list.head.prev
        return sorted_list
        
    while curr:     # Inserting in sorted order
        if movie.title.upper() <= curr.title.upper():
            flag = True
            sorted_list.insert_movie_after(curr.prev, movie)
            break;
        curr = curr.next
    
    if not flag:    
        sorted_list.insert_movie_after(sorted_list.tail, movie)
    
    return sorted_list


#####################################################
# Task 3: complete the function 
# mergeMovieLists(movies1: MovieList, movies2: MovieList) -> MovieList 
#####################################################
def mergeMovieLists(movies1: MovieList, movies2):
    movie = MovieList()
    curr1 = movies1.head    # Pointer for movies1 list
    curr2 = movies2.head    # Pointer for movies2 list
    
    # inserting movies in sorted order
    while curr1 is not None and curr2 is not None:
        
        if curr2.title.upper() <= curr1.title.upper():  # comparing case insenitively
            movie.append_movie(curr2)
            curr2 = curr2.next
        else:
            movie.append_movie(curr1)
            curr1 = curr1.next
    
    # If MovieNodes are left in movie1
    while curr1 is not None:
        movie.append_movie(curr1)
        curr1 = curr1.next
    
    # if MovieNodes are left in movie2
    while curr2 is not None:
        movie.append_movie(curr2)
        curr2 = curr2.next
    
    return movie



###################################################################
# This code is provided as a starting point for your
# development and testing - as you test in DEVELOP MODE
# Please modify the code below as you DEVELOP your program 
# The code in __main__ will not be used to test your program
###################################################################
if __name__ == "__main__":
    
    # Test Task 1
    movies1 = MovieList() 
    movies1.append_movie(MovieNode('The Power of the Dog', 2021, 'Jane Campion','drama'))
    movies1.append_movie(MovieNode('Point Break', 1991,'Kathryn Bigelow', 'action'))
    movies1.append_movie(MovieNode('Point Break', 1991,'Kathryn Bigelow', 'action'))
    movies1.append_movie(MovieNode('The Power of the Dog', 2021, 'Jane Campion','drama'))
    movies1.append_movie(MovieNode('Little Women',2019,'Greta Gerwig','drama'))
    
    movies1 = remove_duplicates(movies1)
    print(f'Task1:\n{movies1}')
    
    
    # Test Task 2    
    movies2 = MovieList()
    movies2.append_movie(MovieNode('Little Women',2019,'Greta Gerwig','drama'))
    movies2.append_movie(MovieNode('The Hurt Locker',2008,'Kathryn Bigelow','action'))
    movies2.append_movie(MovieNode('The Power of the Dog',2021,'Jane Campion','drama'))

    movies2 = addMovieSortedOrder(movies2, MovieNode('Hidden Figures', 2016, 'Theodore Melfi', 'drama'))
    print(f'Task2\n{movies2}')


    # Test Task 3
    merged = mergeMovieLists(movies1, movies2)
    print(f'Task3\n{merged}')
```
